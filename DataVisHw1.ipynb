```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv("./data/west_before_taxes.csv", sep="\t")
df2 = pd.read_csv("./data/ne_before_taxes.csv", sep="\t")
df2.head()
```

```python
x=df['Year']
y1=df.iloc[:,1]
y2=df.iloc[:,2]
y3=df.iloc[:,3]
y4=df.iloc[:,4]
plt.plot(x, y1, marker="s", label="Los Angeles")
plt.plot(x, y2, marker="^", label="San Franscisco")
plt.plot(x, y3, marker="D", label="San Diego")
plt.plot(x, y4, marker="o", label="Seattle")
plt.xlabel("Year")
plt.ylabel("Dollars")
plt.title("Median Income Before Taxes In The West")
ylab=[str(i) + "k" for i in np.arange(55,130,5)]
plt.yticks(np.arange(55000,130000,5000), ylab)
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
x=df2['Year']
y1=df2.iloc[:,1]
y2=df2.iloc[:,2]
plt.plot(x, y1, marker="s", label="New York")
plt.plot(x, y2, marker="^", label="Philadelphia")
plt.xlabel("Year")
plt.ylabel("Dollars")
plt.title("Median Income Before Taxes In The Northeast")
ylab=[str(i) + "k" for i in np.arange(60,110,5)]
plt.yticks(np.arange(60000,110000,5000), ylab)
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
x=df['Year']
y1=df.iloc[:,1]
y2=df.iloc[:,2]
plt.plot(x, y2, c='#991111', marker="h", label="San Franscisco")
plt.plot(x, y1, c='#ff4444', marker="o", label="Los Angeles")
x=df2['Year']
y1=df2.iloc[:,1]
y2=df2.iloc[:,2]
plt.plot(x, y1, c='#4444ff', marker="s", label="New York")
plt.plot(x, y2, c='#111199', marker="D", label="Philadelphia")
plt.xlabel("Year")
plt.ylabel("Dollars")
plt.title("Median Income Before Taxes East Versus West")
ylab=[str(i) + "k" for i in np.arange(55,130,5)]
plt.yticks(np.arange(55000,130000,5000), ylab)
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
fw = pd.read_csv("./data/food_west.csv", sep="\t")
fne = pd.read_csv("./data/food_ne.csv", sep="\t")
x = fw['Year']
y1 = fw.iloc[:,1];
y2 = fne.iloc[:,1];

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
ax1.plot(x, y1, c="b", label="West")
ax1.plot(x, y2, c="r", label="Northeast")
ax1.legend()
ax1.set_ylabel("Dollars")
ax1.set_title("Food Expenditures per Region")
ylab=[str(i) + "k" for i in np.arange(60,91,5)]
ax1.set_yticklabels(ylab)
ax1.set_yticks(np.arange(6000,9100,500))

ax2.set_ylabel("Dollars")
ax2.fill_between(x, y1, y2, where = y2 >= y1, facecolor="red", interpolate=True)
ax2.fill_between(x, y1, y2, where = y2 < y1, facecolor="blue", interpolate=True)
ylab=[str(i) + "k" for i in np.arange(60,91,5)]
ax2.set_yticklabels(ylab)
ax2.set_yticks(np.arange(6000,9100,500))
plt.show()
```
