```python
westal = pd.read_csv("./data/alcohol_west.csv", sep="\t")
norteal= pd.read_csv("./data/alcohol_northeast.csv", sep="\t")
wdd = pd.read_csv("./data/average_expenditures_west.csv", sep="\t")
nedd = pd.read_csv("./data/average_expenditures_northeast.csv", sep="\t")

x1 = westal.iloc[:,1]
x2 = westal.iloc[:,2]
x3 = norteal.iloc[:,1]
x4 = norteal.iloc[:,2]

y1 = wdd.iloc[:,1]
y2 = wdd.iloc[:,2]
y3 = nedd.iloc[:,1]
y4 = nedd.iloc[:,2]

A = np.vstack([x2, np.ones(len(x2))]).T
m, c = np.linalg.lstsq(A, y2, rcond=None)[0]
plt.plot(x2, m*x2 + c, c='#4444ff')

plt.scatter(x1, y1, c='#111199', marker='o', label="Los Angeles")
plt.scatter(x2, y2, c='#4444ff', marker='h', s=100, label="San Francisco")
plt.scatter(x3, y3, c='#991111', marker="s", label="New York")
plt.scatter(x4, y4, c='#ff4444', marker="D", label="Philadelphia")
plt.xlabel("Alcohol Expenditures (Dollars)")
plt.ylabel("Average Expenditures (Dollars)")
ylab=[str(i) + "k" for i in np.arange(55,130,5)]
plt.yticks(np.arange(000,80000,5000), ylab)
plt.legend(loc="best")
plt.title("Average Expenditures Versus Alcohol Expenditures At Each Year")
plt.show()
```
