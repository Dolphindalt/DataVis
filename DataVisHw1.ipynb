```python
westal = pd.read_csv("./data/food_west1.csv", sep="\t")
norteal= pd.read_csv("./data/food_northeast1.csv", sep="\t")
wdd = pd.read_csv("./data/alcohol_west.csv", sep="\t")
nedd = pd.read_csv("./data/alcohol_northeast.csv", sep="\t")

y1 = wdd.iloc[:,1] / westal.iloc[:,1]
y2 = wdd.iloc[:,2] / westal.iloc[:,2]
y3 = nedd.iloc[:,1] / norteal.iloc[:,1]
y4 = nedd.iloc[:,1] / norteal.iloc[:,2]
y5 = wdd.iloc[:,3] / westal.iloc[:,3] # San Diego
y6 = wdd.iloc[:,4] / westal.iloc[:,4] # Seattle

x = wdd['Year']

fig = plt.figure()
ax = fig.add_subplot(111)
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

# Turn off axis lines and ticks of the big subplot
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['right'].set_color('none')
ax.tick_params(labelcolor='w', top=False, bottom=False, left=False, right=False)

ax2.plot(x, y1, c='#111199', label="Los Angeles", linewidth=2.5)
ax1.plot(x, y5, c='#5555ff', label="San Diego", linewidth=2.5)
ax2.plot(x, y3, c='#991111', label="New York", linewidth=2.5)
ax2.plot(x, y4, c='#ff4444', label="Philadelphia", linewidth=2.5)
ax1.plot(x, y6, c='#dd77dd', label="Seattle")
ax1.plot(x, y2, c='#991199', label="San Francisco", linewidth=2.5)

ax1.grid()
ax1_ylab = [str(int(i*100)) + "%" for i in np.arange(0.070, 0.170, 0.02)]
ax1.set_yticks(np.arange(0.07, 0.170, 0.02))
ax1.set_yticklabels(ax1_ylab)

ax2.grid()
ax2_ylab = [str(int(i*100)) + "%" for i in np.arange(0.05, 0.10, 0.01)]
ax2.set_yticks(np.arange(0.05, 0.10, 0.01))
ax2.set_yticklabels(ax2_ylab)

ax2.set_xticks(np.arange(2004,2020,2))
ax1.set_xticks(np.arange(2004,2020,2))

ax.set_ylabel("Percent of Alcoholic Expenditures")
ax.set_xlabel("Year")

h1,l1=ax1.get_legend_handles_labels()
h2,l2=ax2.get_legend_handles_labels()
h = h1+h2
l = l1+l2

box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(h, l, loc='center left', bbox_to_anchor=(1.30, 0.5), fancybox=True)

ax1.set_title("Alcoholic Expenditures as a Percent of Total Food Expenditures")
plt.show()
```

```python
westal = pd.read_csv("./data/alcohol_west.csv", sep="\t")
norteal= pd.read_csv("./data/alcohol_northeast.csv", sep="\t")
wdd = pd.read_csv("./data/west_before_taxes.csv", sep="\t")
nedd = pd.read_csv("./data/ne_before_taxes.csv", sep="\t")

x1 = westal.iloc[:,1]
x2 = westal.iloc[:,2]
x3 = norteal.iloc[:,1]
x4 = norteal.iloc[:,2]
x5 = westal.iloc[:,3]
x6 = westal.iloc[:,4]

y1 = wdd.iloc[:,1]
y2 = wdd.iloc[:,2]
y3 = nedd.iloc[:,1]
y4 = nedd.iloc[:,2]
y5 = wdd.iloc[:,3]
y6 = wdd.iloc[:,4]

fig = plt.figure()
ax = fig.add_subplot(111)

plt.scatter(x2, y2, c='#FF00FF', marker='h', label="San Francisco")
plt.scatter(x5, y5, c='#990099', marker="p", label="San Diego")
plt.scatter(x6, y6, c='#F088F0', marker="P", label="Seattle")
plt.scatter(x1, y1, c='#006400', marker='o', label="Los Angeles")
plt.scatter(x3, y3, c='#00FF00', marker="s", label="New York")
plt.scatter(x4, y4, c='#00FF7F', marker="D", label="Philadelphia")

y5.loc[y5.index.max()+1] = 60000
x5.loc[x5.index.max()+1] = 400

A = np.vstack([x5, np.ones(len(x5))]).T
m, c = np.linalg.lstsq(A, y5, rcond=None)[0]
plt.plot(x5, (m+10)*x5 + c+1000.0, c='#000000', linewidth=2.0)

plt.xlabel("Alcohol Expenditures (Dollars)")
plt.ylabel("Income Before Taxes (Dollars)")
ylab=[str(i) + "k" for i in np.arange(40,135,10)]
plt.yticks(np.arange(40000,135000,10000), ylab)
plt.xticks(np.arange(300, 1300, 100))
plt.legend(loc="best")
plt.title("Income Before Taxes By Alcohol Expenditures Each Year")
plt.savefig('selected.png', dpi=500)
plt.show()
```
